```python
# verificating the ID 
invalid_ids = df[
    ~df["Transaction ID"].str.match(r"^TXN_\d+$", na=False)
]
invalid_ids
```

```python
# verificanting the Transaction Date
# We will check for non-null values that do not match the expected date format (YYYY-MM-DD) and also exclude known invalid entries like "ERROR", "UNKNOWN", and "UNKOWN".
mask = (
    df["Transaction Date"].notna()
    &
    ~df["Transaction Date"].isin(
        ["ERROR", "UNKNOWN", "UNKOWN"]
    )
)

invalid_dates = df[
    mask &
    ~df["Transaction Date"].str.match(
        r"^\d{4}-\d{2}-\d{2}$",
        na=False
    )
]

invalid_dates
```

As we can see, there's no impossible Datas to fix.
