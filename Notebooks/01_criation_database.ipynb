# Creation of a sample from the ENEM dataset

```python
import pandas as pd


tamanho_amostra = 100000


amostras = []


for chunk in pd.read_csv(
    "../DADOS/MICRODADOS_ENEM_2020.csv",
    sep=';',
    encoding='latin1',
    chunksize=50000,
    low_memory=False
):
    
    
    amostra_chunk = chunk.sample(frac=0.1)
    
    amostras.append(amostra_chunk)


df_amostra = pd.concat(amostras)


df_amostra = df_amostra.head(tamanho_amostra)


df_amostra.to_csv(
    "../DADOS/enem_amostra.csv",
    index=False
)
```

# Database creation for SQL analysis

```python
import pandas as pd
import sqlite3

# ler csv
df = pd.read_csv(
    "../DADOS/enem_amostra.csv",
    sep=',',
    encoding='latin1'
)

# criar conexÃ£o
conn = sqlite3.connect("../DADOS/enem.db")

# salvar tabela
df.to_sql(
    "enem",
    conn,
    if_exists="replace",
    index=False
)

conn.close()

```
