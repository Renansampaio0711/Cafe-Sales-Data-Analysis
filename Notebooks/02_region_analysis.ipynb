# Objective of the analysis

We will analyze the influence of students’ geographic region on their ENEM performance.

# Data preparation

```python
import pandas as pd

colunas = [
    "SG_UF_ESC",
    "NU_NOTA_CN",
    "NU_NOTA_CH",
    "NU_NOTA_LC",
    "NU_NOTA_MT",
    "NU_NOTA_REDACAO"
]
df = pd.read_csv(
    "../DADOS/enem_amostra.csv",
    sep=",",
    encoding="latin1",
    usecols=colunas)

print(df.head(5))
```

First, we will remove the rows that contain only NaN values.

```python
df = df.dropna(how="all")
```

Now we will create a new column called "REGION" based on the "SG_UF_ESC" column. To do this, we will use pandas’ map function to map each state to its corresponding region.

```python
regioes = {
    # Norte
    "AC": "Norte",
    "AP": "Norte",
    "AM": "Norte",
    "PA": "Norte",
    "RO": "Norte",
    "RR": "Norte",
    "TO": "Norte",

    # Nordeste
    "AL": "Nordeste",
    "BA": "Nordeste",
    "CE": "Nordeste",
    "MA": "Nordeste",
    "PB": "Nordeste",
    "PE": "Nordeste",
    "PI": "Nordeste",
    "RN": "Nordeste",
    "SE": "Nordeste",

    # Centro-Oeste
    "DF": "Centro-Oeste",
    "GO": "Centro-Oeste",
    "MT": "Centro-Oeste",
    "MS": "Centro-Oeste",

    # Sudeste
    "ES": "Sudeste",
    "MG": "Sudeste",
    "RJ": "Sudeste",
    "SP": "Sudeste",

    # Sul
    "PR": "Sul",
    "RS": "Sul",
    "SC": "Sul"
}

df['REGIÃO'] = df['SG_UF_ESC'].map(regioes)
print(df.head(5))
```

Rows where the REGION value is NaN are not useful for our analysis, so we will remove them.

```python
df = df.dropna(subset=["REGIÃO"])
print(df.head(5))   
```

Creating the column for the average scores.

```python
df['nota_media']= df[["NU_NOTA_CN", "NU_NOTA_CH", "NU_NOTA_LC", "NU_NOTA_MT", "NU_NOTA_REDACAO"]].mean(axis=1)
print(df.head(5))
```

Now we group by region and calculate the mean of the “average_score” column.

```python
media_regiao=df.groupby('REGIÃO')['nota_media'].mean().round(2)
print(media_regiao)
```

It is observed that the Southeast and South regions present the highest average ENEM performance, while the North and Northeast show lower averages. These results suggest the existence of regional inequalities associated with socioeconomic and structural factors in Brazilian education.


# Statistical Analysis
Let’s analyze some statistics for each region.

```python
df = df[df["nota_media"] > 0]
estatisticas = df.groupby("REGIÃO")["nota_media"].agg(
    media="mean",
    mediana="median",
    desvio_padrao="std",
    minimo="min",
    maximo="max",
    quantidade="count"
).round(2).sort_values(by="media", ascending=False)

estatisticas
```

# Conclusion
The statistical analysis of ENEM average scores by region highlights meaningful differences in the performance of Brazilian participants.

The Southeast region showed the highest average performance (576.29) as well as the highest median (574.76), indicating a high and relatively consistent level of achievement among its students. It is followed by the South (569.86) and Central-West (553.30) regions.

The lowest averages were observed in the Northeast (531.71) and North (522.17), suggesting regional disparities in educational performance.

Standard deviations range from approximately 88 to 91 points, indicating significant score dispersion across all regions. The Northeast presented the highest standard deviation (98.94), suggesting greater heterogeneity in student performance, while the South showed the lowest (88.22), indicating more homogeneous scores.

It is also observed that, in all regions, medians are slightly lower than means. This suggests the presence of high-scoring participants influencing regional averages.

The Southeast (6037) and Northeast (5988) regions concentrate the largest number of participants in the analyzed sample.

Overall, the results suggest that socioeconomic factors, educational infrastructure, and regional inequalities may significantly impact students’ performance in the ENEM.
