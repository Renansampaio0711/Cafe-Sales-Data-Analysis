# Data Visualization

Up to this point, we have created a database to build a table with ENEM data.

Now we will analyze the data and relate income to ENEM performance. To do this, we will group the table by income and compute the average of the exam scores for each group.


```python
%%sql

SELECT 
    Q006,
    
    ROUND(
        AVG(
            (
                NU_NOTA_CN +
                NU_NOTA_CH +
                NU_NOTA_LC +
                NU_NOTA_MT +
                NU_NOTA_REDACAO
            ) / 5.0
        ),
        2
    ) AS nota_media

FROM enem

WHERE 
    NU_NOTA_CN IS NOT NULL
    AND NU_NOTA_CH IS NOT NULL
    AND NU_NOTA_LC IS NOT NULL
    AND NU_NOTA_MT IS NOT NULL
    AND NU_NOTA_REDACAO IS NOT NULL

GROUP BY Q006

ORDER BY Q006;
```

As we can see, there is a direct relationship between income and performance.
