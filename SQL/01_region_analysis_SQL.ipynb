Perform a basic analysis of the influence of geographic region on students’ performance in the ENEM.

```python
%%sql

select * 
from enem 
limit 5
```

We will split the table by the regions of Brazil.

```python
%%sql

SELECT 
    regiao,

    ROUND(
        AVG(
            NU_NOTA_MT +
            NU_NOTA_CN +
            NU_NOTA_CH +
            NU_NOTA_LC +
            NU_NOTA_REDACAO
        ) / 5,
        2
    ) AS nota_media

FROM enem_regioes

GROUP BY regiao;

```
